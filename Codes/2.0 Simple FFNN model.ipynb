```python
from keras import layers,models

model=models.Sequential()
# model is the NN
model.add(layers.Dense(3,input_dim=1,activation='linear'))
#1st Hidden Layer
model.add(layers.Dense(1,input_dim=3,activation='linear'))
#output layer
model.compile(loss='mse',optimizer='adam')
```

```python
import numpy as np

x=np.arange(0,1,0.002)
y=(2*np.power(x,2))+(3*x)+4
```

```python
history=model.fit(x,y,epochs=500)
```

```python
result=model.predict([1])
print(result)
```

```python
model.summary()
```

```python
from matplotlib import pyplot as plt

plt.plot(history.history['loss'])
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()
```
